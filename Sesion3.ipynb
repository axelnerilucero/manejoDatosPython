## Generar un DF con de alumnas o alumnos de la tabla alumnos de la base bdcolegio02

```python
from sqlalchemy import create_engine

us = 'fesalu'
passwd = 'bdfes2023'
url = 'bd.arcelia.net'
puerto = '3306'
bd = 'bdcolegio02'
auth = 'mysql_native_password'

str_conn = 'mysql+mysqlconnector://{0}:{1}@{2}:{3}/{4}?auth_plugin={5}'
str_conn = str_conn.format(us, passwd, url, puerto, bd, auth)

str_conn
```

```python
conn = create_engine(str_conn)
conn
```

```python
sql = '''SELECT * FROM alumnos WHERE sexo =  %s'''

df_mysql = pd.read_sql_query(sql=sql, con=conn, params=[('F',)])
```
